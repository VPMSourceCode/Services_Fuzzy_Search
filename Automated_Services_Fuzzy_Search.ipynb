### Fuzzy Search Algorithm

```python
import math
from fuzzywuzzy import fuzz

# Create a new column to store the match results
df2 = df2.copy()
df2['Match'] = None

# Iterate over each row in the data frame
for i, row in df2.iterrows():
    service = row['Service'] if not pd.isna(row['Service']) else row['Vendor']
    if isinstance(service, float) and math.isnan(service):
        df2.at[i, 'Match'] = None
        continue
    service_words = service.split()
    max_ratio = 0
    match = None
    match_found = False
    
    # Do a fuzzy search with the whole service string
    for item in data:
        type_ = item['type']
        ratio = fuzz.token_set_ratio(service, type_)
        
        # If the ratio is 100, an exact match is found
        if ratio == 100:
            match = type_
            match_found = True
            break
        
    # If an exact match is found, store the match and move on to the next service string
    if match_found:
        df2.at[i, 'Match'] = match
        continue
    else:
        # If no exact match is found, reset max_ratio
        max_ratio = 0
    
    # Do a fuzzy search with the first word of the service string
    first_word = service_words[0]
    for item in data:
        type_ = item['type']
        ratio = fuzz.token_set_ratio(first_word, type_)
        
        if ratio > max_ratio:
            max_ratio = ratio
            match = type_
    
    # If a match is found with the first word, keep checking the next words until no match is found or no more words left
    if max_ratio > 0:
        search_string = first_word
        for word in service_words[1:]:
            new_search_string = search_string + ' ' + word
            for item in data:
                type_ = item['type']
                ratio = fuzz.token_set_ratio(new_search_string, type_)
                
                # If a match is found, update the search string and store the match
                if ratio > max_ratio:
                    max_ratio = ratio
                    match = type_
                    search_string = new_search_string
            
            # If no match is found with the new search string, break the loop
            if max_ratio == 0:
                break
    
    # If no match is found with the service string or first word, check each word individually
    if max_ratio == 0:
        for word in service_words:
            for item in data:
                top5_words = [word['word'] for word in item['top5']]
                ratio = fuzz.token_set_ratio(word, ' '.join(top5_words))
                
                # If a match is found,
                if ratio > max_ratio:
                    max_ratio = ratio
                    match = item['type']
                    
    # Choose accuracy
    if max_ratio > 60:
      df2.at[i, 'Match'] = match
    else:
      df2.at[i, 'Match'] = "-"
```
